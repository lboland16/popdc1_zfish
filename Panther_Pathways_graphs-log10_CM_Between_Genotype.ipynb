```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

df3 = pd.read_csv('./CM3_Panther_Pathways.csv')
df3.head(10)
#this is between clusters as there were so few wt in this population

x3 = np.arange(len(df3['Panther Pathways'] ))
plt.rcParams["figure.figsize"] = [10, 15]
df3["log10pvalue"].plot(kind='barh', legend=False, 
                color=(df3['Fold Enrichment'] > 0).map({True: 'yellowgreen',
                                                    False: 'indianred'}))
plt.title("Differentially regulated Panther Pathways in CM3")
plt.xlabel('-log10(p value)')
plt.ylabel("Panther Pathways")
plt.yticks(x3, df3["Panther Pathways"])
plt.savefig('Differentially regulated Panther Pathways in CM3.png',dpi=300, bbox_inches='tight')
