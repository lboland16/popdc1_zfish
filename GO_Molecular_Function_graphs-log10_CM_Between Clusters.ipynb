```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

df3 = pd.read_csv('./CM3_GO_Molecular_Function.csv')
df3.head(10)
#this is between clusters as there were so few wt in this population

x3 = np.arange(len(df3['GO Molecular Function'] ))
plt.rcParams["figure.figsize"] = [10, 15]
df3["log10pvalue"].plot(kind='barh', legend=False, 
                color=(df3['Fold Enrichment'] > 0).map({True: 'yellowgreen',
                                                    False: 'indianred'}))
plt.title("Differentially regulated GO Molecular Function in CM3")
plt.xlabel('-log10(p value)')
plt.ylabel("GO Molecular Function")
plt.yticks(x3, df3["GO Molecular Function"])
plt.savefig('Differentially regulated GO Molecular Function in CM3.png',dpi=300, bbox_inches='tight')
